```python
# 심봤다 테스트
from keras.models import Sequential
from keras.layers.core import Dense
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import StratifiedKFold

import pandas as pd
import numpy as np 
import tensorflow as tf

# seed 값 설정
seed = 42
np.random.seed(seed)
tf.random.set_seed(seed)
```

```python
# 데이터 입력
pd.read_csv('/content/drive/MyDrive/dataset/sonar.csv')
```

```python
dataset = df.values
x = dataset[:,:-1]
x = np.asarray(x).astype(np.float32) # 타입 변경 필요(실수형 float32)
# Failed to convert a NumPy array to a Tensor (Unsupported object type float).
y_obj=dataset[:,-1]
```

```python
x, y_obj
```

```python
# 문자열 변환
le = LabelEncoder()
le.fit(y_obj)
y = le.transform(y_obj)
```

```python
# 10개의 파일로 쪼갬

n_fold=10
skf = StratifiedKFold(n_splits=n_fold, shuffle=True, random_state=seed)
```

```python
accuracy=[ ]

# 모델 설정, 컴파일, 실행

for train, test in skf.split(x, y):
  model = Sequential()
  model.add(Dense(24, input_dim=60, activation ='relu'))
  model.add(Dense(10, activation='relu'))
  model.add(Dense(1,activation='sigmoid'))

  model.compile(loss = 'mean_squared_error', 
                optimizer = 'adam',
                metrics=['accuracy'])
  model.fit(x[train], y[train], epochs = 100, batch_size=5)
  k_acc = '%.4f'%(model.evaluate(x[test],y[test])[1])
  accuracy.append(k_acc)
```

```python
# 결과 출력 
print('\n%.2f fold accuracy :'% n_fold, accuracy)
```
