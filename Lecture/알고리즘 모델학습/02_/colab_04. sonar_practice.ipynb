```python
# 심봤다 테스트
from keras.models import Sequential
from keras.layers.core import Dense
from sklearn.preprocessing import LabelEncoder

import pandas as pd
import numpy as np 
import tensorflow as tf

# seed 값 설정
seed = 42
np.random.seed(seed)
tf.random.set_seed(seed)
```

```python
# 데이터 입력
pd.read_csv('/content/drive/MyDrive/dataset/sonar.csv', header=None)
```

```python
df = pd.read_csv('/content/drive/MyDrive/dataset/sonar.csv', header=None)
```

```python
dataset = df.values
x = dataset[:,:-1]
x = np.asarray(x).astype(np.float32) # 타입 변경 필요(실수형 float32)
# Failed to convert a NumPy array to a Tensor (Unsupported object type float).
y_obj=dataset[:,-1]
```

```python
x, y_obj
```

```python
# 문자열 변환
le = LabelEncoder()
le.fit(y_obj)
y = le.transform(y_obj)
```

```python
from pandas.core.common import random_state
# 학습 셋과 테스트 셋 구분(train_test_split()) 7:3 비율
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3, random_state= seed)
```

```python
# 모델설정
model = Sequential()
model.add(Dense(24, input_dim=60, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='sigmoid')) # 이진분류 : 시그모이드 함수 사용
```

```python
# 모델 컴파일 (모델 환경 설정)

model.compile(loss = 'mean_squared_error',    # 오차함수 : mse
              optimizer = 'adam',
              metrics = ['accuracy'])
```

```python
# 모델 실행 (fit)
# model.fit(x, y, epochs = 120, batch_size = 5) # train_test 나누기 전 (전체)
model.fit(x_train, y_train, epochs=120, batch_size = 5) # train set 사용해야 함
```

```python
# 결과 출력 및 평가

# model.evaluate(x,y) # train_test 구분 전
model.evaluate(x_test, y_test) # evaluate 할 때 : test set
```

```python
print('\n Accuracy : %.4f'%(model.evaluate(x,y)[1]))
```

```python
model.save('sonar_model.h5') # 모델 저장
```
