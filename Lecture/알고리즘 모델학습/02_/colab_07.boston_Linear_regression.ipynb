```python
#-*- coding: utf-8 -*-
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split

import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
# seed 값 설정 

seed = 42
np.random.seed(seed)
tf.random.set_seed(42)
```

```python
pd.read_csv('/content/drive/MyDrive/dataset/housing.csv', delim_whitespace= True, header=None)
# delim_whitespace= True : 공백이 많은 데이터 일때에 공백을 지워주는 함수
```

```python
df = pd.read_csv('/content/drive/MyDrive/dataset/housing.csv', delim_whitespace= True, header=None)
```

```python
df.info()
```

```python
df.head()
```

```python
dataset = df.values
x = dataset[:,:-1 ]
y = dataset[:,-1 ]
```

```python
x.shape
```

```python
x_train, x_test, y_train, y_test = \
      train_test_split(x,y,test_size=0.3,random_state=seed)
```

```python
# 모델 생성
model = Sequential()
model.add(Dense(30, input_dim=13, activation ='relu'))
model.add(Dense(6, activation ='relu'))
model.add(Dense(1)) # 활성화 함수 없음 (회귀 : 예측)
```

```python
# 환경설정
model.compile(loss='mean_squared_error',
              optimizer = 'adam')
```

```python
# 모델학습
model.fit(x_train,y_train, epochs=200, batch_size=10)
```

```python
# 예측값과 실제값 비교

model.predict(x_test)
```

```python
model.predict(x_test).flatten()
```

```python
y_test # 실제값
```

```python
y_pred = model.predict(x_test).flatten() # .flatten() 실제값과 같은 차원으로 맞춰주기위해 사용
y_pred # 예측값
```

```python
y_pred = model.predict(x_test).flatten()

for i in range(10):
  label = y_test[i]
  prediction = y_pred[i]

  print("실제가격: {:.3f},예상가격:{:.3f}".format(label, prediction))
```
